```python
#Importando Lib neccessárias
import pandas as pd
```

```python
#Criando DataFrame
df=pd.read_excel('AcoesBruto.xlsx')
```

```python
#Visualizando status do dataframe
df.head()
```

```python
#Excluindo colunas desnecessárias
df=df.drop(columns=['price.high','price.low','price.adjusted','ticker','ret.adjusted.prices','ret.closing.prices','Empresa','Tipo','Classificação'])
```

```python
#Visualizando status do novo DataFrame
df.head()
```

```python
#Reorganizando colunas
df.loc(axis=1)['ref.date','price.open','price.close','volume','Setor','Subsetor']
```

```python
#renomeando colunas:
df=df.rename(columns={'ref.date':'Data','price.open':'Preço de Entrada','price.close':'Preço de saída'})
```

```python
#Verificando status do Dataframe
df.head()
```

```python
#Filtro de 2020
filtro_2020=(df['Data']>='2020-01-01')&(df['Data']<='2020-12-31')
df=df[filtro_2020]
df.head()
```

```python
#Agrupando Dados
df=df.groupby(['Setor','Subsetor','Data']).sum()
df.head()
```

```python
df.to_excel('Tabelafinal.xlsx')
```
